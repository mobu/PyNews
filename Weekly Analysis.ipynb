```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
breitbart = pd.read_csv('syria/Breitbart/Stats Breitbart.csv')
cnn = pd.read_csv('syria/CNN/Stats CNN.csv')
bbc = pd.read_csv('syria/BBC News/Stats BBC.csv')
```

```python
plt.figure(figsize=(20,15))
sns.set(color_codes=True)
sns.regplot(x='Subjectivity',y='Polarity',data=breitbart,label='Breitbart')
sns.regplot(x='Subjectivity',y='Polarity',data=cnn,label='CNN')
sns.regplot(x='Subjectivity',y='Polarity',data=bbc,label='BBC')
plt.legend()
```

```python
plt.figure(figsize=(20,15))
sns.set(color_codes=True)
sns.kdeplot(breitbart['Polarity'],shade=True,color='red',label='Breitbart')
sns.kdeplot(cnn['Polarity'],shade=True,color='olive',label='CNN')
sns.kdeplot(bbc['Polarity'],shade=True,color='skyblue',label='BBC')
plt.legend()
```

```python
plt.figure(figsize=(20,15))
sns.set(color_codes=True)
sns.lmplot(x='Polarity',y='Subjectivity',hue='Date',size=5,data=breitbart)
plt.legend()
```
